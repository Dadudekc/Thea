
# ðŸ“ˆ Backtesting Notebook (`backtest.ipynb`)

This Jupyter Notebook allows for **backtesting trading strategies** using the `Backtester` class.
It loads historical data, applies a trading strategy, and visualizes the results.

### âœ… Features:
- **Runs backtests** using custom strategies.
- **Calculates trading signals** (BUY/SELL/HOLD).
- **Visualizes performance** with price charts & cumulative returns.
- **Flexible**â€”easily swap out different trading strategies.



```python
# âœ… Ensure correct imports
import sys

sys.path.append(r"D:\TradingRobotPlug")  # Manually add the project path if necessary


import pandas as pd

# âœ… Import core modules
from basicbot.backtester import Backtester
from basicbot.logger import setup_logging
from basicbot.strategy import Strategy

# âœ… Setup Logger
logger = setup_logging("BacktestNotebook")

# âœ… Initialize Strategy
strategy = Strategy()

# âœ… Initialize Backtester
backtester = Backtester(strategy=strategy, logger=logger)

# âœ… Load Sample Historical Data (Ensure DataFrame Contains 'close' Column)
df = pd.DataFrame({
    "close": [100, 102, 104, 103, 105, 107, 106, 108, 110, 112],
    "high":  [101, 103, 105, 104, 106, 108, 107, 109, 111, 113],
    "low":   [99, 101, 103, 102, 104, 106, 105, 107, 109, 111]
})

# âœ… Run the Backtest
logger.info("ðŸš€ Running Backtest...")
results = backtester.run_backtest(df)

# âœ… Display Backtest Results
print("\nðŸ“Š Backtest Results:")
print(results)

# âœ… Save Logs (If Needed)
logger.info("âœ… Backtest Completed.")

# âœ… Visualize Backtest Performance (Optional)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(results["cumulative_returns"], label="Cumulative Strategy Returns", linestyle="--", color="blue")
plt.xlabel("Time")
plt.ylabel("Cumulative Returns")
plt.title("Backtest Performance")
plt.legend()
plt.grid(True)
plt.show()
```
